```python
#Pie Chart 

expenses=["raw materials","salary","transport","manufacturing","Extra"]
amount=[150,50,80,20,100]
colors=["red","pink","blue","yellow","orange"]
explode=[0,0,0,0.4,0]
plt.pie(amount,labels=expenses,autopct="%1.1f%%",colors=colors,
        wedgeprops={
            "edgecolor":"black",
            "linewidth":1,
            "linestyle":"--"

        },
        explode=explode
        )

```
