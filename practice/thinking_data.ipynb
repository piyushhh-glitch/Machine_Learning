```python
import json
```

```python
def load_data(filename):
    with open(filename,"r+") as f:
        data=json.load(f)
    return data
    
```

```python
data=load_data("store_data.json")
print(data)
```

```python
print(type(data))
```

```python
def clean_data(data):

    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_data=[]
    unique_data=set()

    for user in data:

    

        raw_rating=str(user["rating"]).strip().lower()

        if (raw_rating in text_to_num):
            raw_rating=text_to_num[raw_rating]

        user["rating"]=raw_rating


        raw_age=user.get("age")
        if raw_age==None:
            user["age"]=None

        if (user["name"].strip() in unique_data):
            continue
        unique_data.add(user["name"])
        cleaned_data.append(user)

    return cleaned_data
```

```python
clean_data(data)

```

```python
def get_insights(data):

    total_rating=0

    for user in data:
        total_rating+=float(user["rating"])

    poor_rating=0
    for user in data:
        if float(user["rating"])<3:
            poor_rating+=1


    print(f"avg rating is ={total_rating//len(data)}")
    print(f"percentage of poor rating is={poor_rating/len(data)*100}%")
```

```python
get_insights(data)
```

```python
def get_recommendations(data):
    recommendation=[]

    for user in data:
        curr_recomm={}
        curr_recomm["name"]=user["name"]

        if float(user["rating"])<=4:
            curr_recomm["brand"]="apple"
        else:
            curr_recomm["brand"]="samsung"

        recommendation.append(curr_recomm)
    return recommendation
```

```python
get_recommendations(data)
```
