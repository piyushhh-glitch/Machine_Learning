```python
# RESHAPING METHODS
#1.Melt-convert a wide format dataframe into long/tidy format

df=pd.DataFrame({
    "Country":["USA" ,"USA", "INDIA" ,"INDIA"],
    "year":[2021 ,2022, 2021 ,2022],
    "sales":[100 ,120 ,90 ,110],
    "profit":[20,25,18,22]
})

melted=df.melt(
    id_vars=["Country","year"],
    value_vars=["sales","profit"],
    var_name="metrics",
    value_name="values"
)
melted
```

```python
# 2.Convert a long format data into a wide format
original=melted.pivot(
    index=["Country","year"],
    columns="metrics",
    values="values"
)
original
```

Combining and Joining methods

```python
df_customers=pd.DataFrame({
    "customer_id":[1,2,3,4],
    "name":["adam","eve","bob","charlie"]
})

df_orders=pd.DataFrame({
    "order_id":[122,101,122,134],
    "customer_id":[2,3,1,4],
    "amount":[250,120,780,670]
})

pd.merge(df_customers,df_orders,on="customer_id")   #Inner join
pd.merge(df_customers,df_orders,on="customer_id",how="left")   #left join
pd.merge(df_customers,df_orders,on="customer_id",how="right")   #right join
pd.merge(df_customers,df_orders,on="customer_id",how="outer")   #outer join
```
