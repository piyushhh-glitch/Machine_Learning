```python
import requests
from bs4 import BeautifulSoup
```

```python
#Fetch data and store it 

page_count=1

while True:
    URL=f"https://quotes.toscrape.com/page/{page_count}/"
    res=requests.get(URL)

    soup=BeautifulSoup(res.text,"lxml")
    quotes=soup.select("div.quote")

    if not quotes:
        print("No Valid Pages Anymore")
        break

    with open(f"scrap_data/quotes{page_count}.html","w") as f:
        f.write(res.text)
        print(f"downloaded data from page{page_count}")

    page_count+=1
```

```python
#Extract UseFul Info
with open("scrap_data/quotes1.html","r") as f:
    html_content=f.read()

soup=BeautifulSoup(html_content,"lxml")


```

```python
all_quotes=soup.select("div.quote")
life_quotes=[]

for q in all_quotes:
    all_tags=[]

    for tag in q.select(".tags .tag"):
        all_tags.append(tag.get_text())


    

    if "life" in all_tags:
        text=q.select_one("span.text").get_text()
        author=q.select_one("small.author").get_text()
        life_quotes.append([text,author])
print(life_quotes)
```

```python
scrap_count=1
life_quotes=[]

while True:
    try:
        with open(f"scrap_data/quotes{scrap_count}.html","r") as f:
            html_content=f.read()

        soup=BeautifulSoup(html_content,"lxml")
        all_quotes=soup.select("div.quote")
        
        for q in all_quotes:
            all_tags=[]

            for tag in q.select(".tags .tag"):
                all_tags.append(tag.get_text())

            if "life" in all_tags:
                text=q.select_one("span.text").get_text()
                author=q.select_one("small.author").get_text()
                life_quotes.append([text,author])

        scrap_count+=1
    except FileNotFoundError:
        break
```
